```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 讀取球隊資料
men_teams = pd.read_csv("../../data/1The Basics/MTeams.csv")
women_teams = pd.read_csv("../../data/1The Basics/WTeams.csv")

# 顯示男子球隊資料
print("Men's Teams:")
print(men_teams.head())

# 顯示女子球隊資料
print("\nWomen's Teams:")
print(women_teams.head())

# 球隊數量
print("\nNumber of Men's Teams:", len(men_teams))
print("Number of Women's Teams:", len(women_teams))

```

```python

# 讀取 MSeasons.csv 和 WSeasons.csv 文件
men_seasons = pd.read_csv("../../data/1The Basics/MSeasons.csv")
women_seasons = pd.read_csv("../../data/1The Basics/WSeasons.csv")

# 分析賽季數量
men_season_count = men_seasons['Season'].nunique()
women_season_count = women_seasons['Season'].nunique()
print("\nNumber of Men's Seasons:", men_season_count)
print("Number of Women's Seasons:", women_season_count)

# 分析區域分佈
men_regions = men_seasons[['RegionW', 'RegionX', 'RegionY', 'RegionZ']].stack().value_counts()
women_regions = women_seasons[['RegionW', 'RegionX', 'RegionY', 'RegionZ']].stack().value_counts()
print("\nMen's Region Distribution:")
print(men_regions)
print("\nWomen's Region Distribution:")
print(women_regions)
```
